# ACL 2021

```python
with open('ACL2021.txt', 'r') as f:
    data = f.read()
    data = data.split('Main Conference')[1].split('Findings of ACL')
    main_papers = data[0].strip().split('\n\n')
    findings_papers = data[1].strip().split('\n\n')
```

```python
text = ''
for line in main_papers + findings_papers:
    if line != 'Long Papers' and line != 'Short Papers':
        tmp = line.split('\n')[0]
        text += f'\n{tmp}'
```

```python
# Display the generated image:
from wordcloud import WordCloud
import matplotlib.pyplot as plt

STOPWORDS = list(map(str.strip, open('stopwords').readlines()))
STOPWORDS += ['Using', 'via', 'based', 'Improving', 'Improve', 'Learning', 'Word', 
              'Model', 'Models', 'Text', 'Task', 'Tasks', 'Data', 'Language', 'Languages',
              'Modeling', 'Toward', 'Towards', 'Multi', 'Enhanced', 
             'Training', 'Approach', 'Network', 'Networks', 'Human', 
              'Natural', 'Neural', 'Level', 'Pre', 'End', 'Aware', 'New', 
             'trained', 'Deep', 'Label', 'Exploring', 'Better', 'NLP', 
              'Enhancing', 'Classification', 'Method', 's', 'Generation', 
             'BERT', 'Attention', 'Generating', 'Embedding', 'Embeddings', 
              'Joint', 'Aspect', 'Code', 'Prediction', 'Use', 'Dataset', 'Type', 'Words', 
             'Large Scale', 'Table', 'Guided', 'Framework', 'Resource', 'Oriented', 'Need', 
             'News', 'Simple', 'English', 'Impact', 'Texts', 'Response', 'Beyond']

# lower max_font_size
wordcloud = WordCloud(width=700, height=350, max_font_size=36, min_font_size=4, stopwords=set(STOPWORDS), mode='RGBA', colormap='tab10', background_color=None).generate(text)
plt.figure(figsize=(40, 20))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
wordcloud_svg = wordcloud.to_svg(embed_font=True)
f = open("ACL2021.svg","w+")
f.write(wordcloud_svg)
f.close()
```

# IJCAI 2021

```python
with open('IJCAI2021.txt', 'r') as f:
    data = f.read()
    data = data.split('\n')
```

```python
text = ''
for line in data:
    if line[0] == '#':
        text += f'\n{line}'
```

```python
# Display the generated image:
from wordcloud import WordCloud
import matplotlib.pyplot as plt

STOPWORDS = list(map(str.strip, open('stopwords').readlines()))
STOPWORDS += ['Model', 'Neural Network', 'Neural Networks', 'Network', 'Networks', 'Attention', 
              'Data', 'Modeling', 'based', 'Using', 'Learning', 'Toward', 'Matching', 
              'Multi', 'via', 'Deep', 'Label', 'Aware', 'Problem', 'Neural', 'Towards', 
             'Algorithm', 'Prediction', 'Framework', 'Two', 'Approach', 'Classification', 
             'Task', 'Models', 'Solving', 'Method', 'Improving', 't', 'One', 'Tasks', 
             'User', 'Algorithms', 'New', 'Combining', 'High', 'End']

# lower max_font_size
wordcloud = WordCloud(width=700, height=350, max_font_size=36, min_font_size=4, stopwords=set(STOPWORDS), mode='RGBA', colormap='tab10', background_color=None).generate(text)
plt.figure(figsize=(40, 20))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
wordcloud_svg = wordcloud.to_svg(embed_font=True)
f = open("IJCAI2021.svg","w+")
f.write(wordcloud_svg )
f.close()
```
